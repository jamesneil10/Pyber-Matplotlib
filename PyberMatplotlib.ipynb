```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Incorporate Seaborn if preferred 
import seaborn as sns

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)


```

```python
ride_data.head()
```

```python
city_fares = ride_data.groupby('city')['fare'].agg(['mean','count'])
city_fares.head()
```

```python
# Combine the data into a single dataset
pyber = pd.merge(city_fares, city_data, on='city')
pyber = pyber.rename(columns={'city':'City', 'mean':'Average Fare', 'count':'Total Number of Rides',
                              'driver_count':'Total Number of Drivers', 'type':'City Type'})
pyber.head()
```

```python
# Combine the data into a single dataset
pyber_merged = pd.merge(ride_data, city_data,on='city')
pyber_merged
```

```python
Bubble Plot of Ride Sharing Data
```

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types
plt.scatter(pyber["Total Number of Rides"][pyber["City Type"]=="Urban"],
           pyber["Average Fare"][pyber["City Type"]=="Urban"],
           marker='o',
            color= '#ff7f50',
            edgecolors = 'black',
            s = pyber["Total Number of Drivers"][pyber["City Type"]=="Urban"]*12,
            label="Urban"
           )
plt.scatter(pyber["Total Number of Rides"][pyber["City Type"]=="Suburban"],
           pyber["Average Fare"][pyber["City Type"]=="Suburban"],
           marker='o',
            color= '#00ccff',
            edgecolors = 'black',
            s = pyber["Total Number of Drivers"][pyber["City Type"]=="Suburban"]*12,
            label="Suburban"
           )
plt.scatter(pyber["Total Number of Rides"][pyber["City Type"]=="Rural"],
           pyber["Average Fare"][pyber["City Type"]=="Rural"],
           marker='o',
            color= '#ffd700',
            edgecolors = 'black',
            s = pyber["Total Number of Drivers"][pyber["City Type"]=="Rural"]*12,
            label="Rural"
           )

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)

# Create a legend
plt.legend()
# Incorporate a text label regarding circle size
plt.text(45,35,"Note:\n Circle size correlates with driver count per city.")
# Save Figure
plt.savefig("PyberRideSharingData.png")
```

```python
plt.show()
```

```python
total_fare = pyber_merged['fare'].sum()
total_rides = pyber_merged['fare'].count()
total_drivers = city_data['driver_count'].sum()
colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,.1)
```

```python
Total Fares by City Type
```

```python
# Calculate Type Percents
percent_fare = pyber_merged.groupby('type')['fare'].agg(['sum']).reset_index()
percent_fare['% of Total Fares'] = percent_fare['sum']/total_fare

# Build Pie Chart
plt.pie(percent_fare["% of Total Fares"], explode=explode, labels=percent_fare["type"], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("pctTotalFares.png")

# Show Figure
plt.show()
```

Total Rides by City Type

```python
# Calculate Ride Percents
percent_rides = pyber_merged.groupby('type')['fare'].agg(['count']).reset_index()
percent_rides['% of Total Rides'] = percent_rides['count']/total_rides

# Build Pie Chart
plt.pie(percent_rides["% of Total Rides"], explode=explode, labels=percent_rides["type"], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("pctTotalRides.png")

# Show Figure
plt.show()
```

Total Drivers by City Type

```python
# Calculate Driver Percents
percent_drivers = city_data.groupby('type')['driver_count'].agg(['sum']).reset_index()
percent_drivers['% of Total Drivers'] = percent_drivers['sum']/total_drivers

# Build Pie Charts
plt.pie(percent_drivers["% of Total Drivers"], explode=explode, labels=percent_drivers["type"], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("pctTotalDrivers.png")

# Show Figure
plt.show()
```

Three observable trends of data:
    1)There are significantly more ride share drivers in urban areas compared to suburban or rural areas.
    2)A reason for the saturation of drivers in urban areas is because there is significantly more rides in urban areas compared      to suburban or rural.
    3)Urban areas account for 62.7% of fares compared to 30.5%(suburban) and 6.8%(rural).
